```python
# thêm các thư viện cần thiết
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.linear_model import LinearRegression, Lasso
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
from sklearn.preprocessing import  OrdinalEncoder, StandardScaler
from sklearn.ensemble import StackingRegressor
```

```python
#thêm dữ liệu
data = pd.read_csv('./data/student_performance.csv')
data
```

```python
#Bỏ cột dữ liệu không cần thiết
data.drop(columns=['StudentID', 'Name', 'Gender'], inplace=True)

list = [['Low', 'Medium', 'High']]

PL = OrdinalEncoder(categories=list)
data['ParentalSupport'] = PL.fit_transform(data[['ParentalSupport']])

data.head()
```

```python
#Chia dữ liệu và tiền xử lý dữ liệu
X = data.iloc[:,:-1]
y = data.iloc[:,-1]

#Xử lý dữ liệu
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# chia dữ liệu thành tập huấn luyện và tập kiểm tra
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=42)
# # Giả sử y_train chứa các giá trị liên tục, cần chuyển nó về nhị phân
# threshold = 0.5
# y_train_bool = (y_train >= threshold).astype(int)
# y_test_bool = (y_test >= threshold).astype(int)

X_train
```

Xây dựng mô hình bằng phương pháp Neural network 

```python
# Xây dựng mô hình neural network bằng thư viện MlpRegressor

mlp_model_neural = MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=3000, random_state=0)

mlp_model_neural.fit(X_train, y_train)

y_pred_neural = mlp_model_neural.predict(X_test)


mae_neural = mean_absolute_error(y_test, y_pred_neural)
r2_neural = r2_score(y_test, y_pred_neural)
mse_neural = mean_squared_error(y_test, y_pred_neural)
rmse_neural = np.sqrt(mse_neural)
```

```python
# In ra các giá trị đánh giá mô hình
print(f"Mean Absolute Error (neural): {mae_neural}")
print(f"R2_score (neural): {r2_neural}")
print(f"Mean Squared Error (neural): {mse_neural}")
print(f"Root Mean Squared Error (neural): {rmse_neural}")
```

Xây dựng mô hình bằng phương pháp hồi quy tuyến tính

```python
# Xây dựng mô hình với phương pháp hồi quy tuyến tính
regressor = LinearRegression()
regressor.fit(X_train, y_train)

y_prep_linear = regressor.predict(X_test)

mae_linear = mean_absolute_error(y_test, y_prep_linear)

r2_linear = r2_score(y_test, y_prep_linear)

mse_linear = mean_squared_error(y_test, y_prep_linear)

rmse_linear = np.sqrt(mse_linear)
```

```python
# In ra các giá trị đánh giá mô hình
print(f"Mean Absolute Error (linear): {mae_linear}")
print(f"R2_score (linear): {r2_linear}")
print(f"Mean Squared Error (linear): {mse_linear}")
print(f"Root Mean Squared Error (linear): {rmse_linear}")
```

Xây dựng mô hình bằng phương pháp Lasso

```python
lasso_model = Lasso(alpha=0.1)
lasso_model.fit(X_train, y_train)

y_pred_lasso = lasso_model.predict(X_test)

mae_lasso = mean_absolute_error(y_test, y_pred_lasso)

r2_lasso = r2_score(y_test, y_pred_lasso)

mse_lasso = mean_squared_error(y_test, y_pred_lasso)

rmse_lasso = np.sqrt(mse_lasso)
```

```python
print(f"Mean Absolute Error (lasso): {mae_lasso}")
print(f"R2_score (lasso): {r2_lasso}")
print(f"Mean Squared Error (lasso): {mse_lasso}")
print(f"Root Mean Squared Error (lasso): {rmse_lasso}")
```

Kết hợp các mô hình bằng phương pháp Stacking

```python
#Tạo các base model
base_models = [
    ('neural', MLPRegressor(hidden_layer_sizes=(100, 50), max_iter=3000, random_state=0)),
    ('linear', LinearRegression()),
    ('lasso', Lasso(alpha=0.1))
]

#Tạo mô hình stacking
Stacking_model = StackingRegressor(
    estimators=base_models,
    final_estimator=LinearRegression()
)

#Huấn luyện mô hình
Stacking_model.fit(X_train, y_train)

#Dự đoán giá trị
y_pred_stacking = Stacking_model.predict(X_test)

#Đánh giá mô hình
mae_stacking = mean_absolute_error(y_test, y_pred_stacking)
n2_stacking = r2_score(y_test, y_pred_stacking)
mse_stacking = mean_squared_error(y_test, y_pred_stacking)
rmse_stacking = np.sqrt(mse_stacking)

#In ra các giá trị đánh giá mô hình
print(f"Mean Absolute Error (stacking): {mae_stacking}")
print(f"R2_score (stacking): {n2_stacking}")
print(f"Mean Squared Error (stacking): {mse_stacking}")
print(f"Root Mean Squared Error (stacking): {rmse_stacking}")
```
